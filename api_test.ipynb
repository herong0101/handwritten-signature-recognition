### Gemini model comparison text

```python
import os
from dotenv import load_dotenv
from google import genai
from google.genai import types

load_dotenv()

api_key = os.getenv("GEMINI_API_KEY")
if not api_key:
    raise ValueError("找不到 GEMINI_API_KEY，請先確認 .env 已設定")

client = genai.Client(api_key=api_key)

models = [
    "gemini-3.1-flash-lite-preview",
    "gemini-3-flash-preview",
]

prompt = "說個笑話"
temperature = 0.8
max_output_tokens = 256
top_p = 0.95

def extract_text(response):
    candidates = getattr(response, "candidates", None) or []
    if not candidates:
        return ""
    parts = getattr(candidates[0].content, "parts", None) or []
    return "".join(part.text for part in parts if getattr(part, "text", None)).strip()

for model in models:
    print(f"=== {model} ===")
    try:
        response = client.models.generate_content(
            model=model,
            contents=prompt,
            config=types.GenerateContentConfig(
                temperature=temperature,
                max_output_tokens=max_output_tokens,
                top_p=top_p,
            ),
        )
        print(extract_text(response))
    except Exception as exc:
        print(f"Error: {exc}")
    print()
```

### Gemini model comparison image

```python
import os
from pathlib import Path

from dotenv import load_dotenv
from google import genai
from google.genai import types

load_dotenv()

api_key = os.getenv("GEMINI_API_KEY")
if not api_key:
    raise ValueError("找不到 GEMINI_API_KEY，請先確認 .env 已設定")

image_path = Path("test.png")
if not image_path.exists():
    raise FileNotFoundError("找不到 test.png，請先放一張要測試的圖片在專案根目錄")

image_bytes = image_path.read_bytes()
image_part = types.Part.from_bytes(data=image_bytes, mime_type="image/png")

client = genai.Client(api_key=api_key)

models = [
    "gemini-3.1-flash-lite-preview",
    "gemini-3-flash-preview",
]

prompt = "請用繁體中文簡短總結這張圖片的內容。"
temperature = 0.2
max_output_tokens = 4096
top_p = 0.95

def extract_text(response):
    candidates = getattr(response, "candidates", None) or []
    if not candidates:
        return ""
    parts = getattr(candidates[0].content, "parts", None) or []
    return "".join(part.text for part in parts if getattr(part, "text", None)).strip()

for model in models:
    print(f"=== {model} ===")
    try:
        response = client.models.generate_content(
            model=model,
            contents=[prompt, image_part],
            config=types.GenerateContentConfig(
                temperature=temperature,
                max_output_tokens=max_output_tokens,
                top_p=top_p,
            ),
        )
        print(extract_text(response))
    except Exception as exc:
        print(f"Error: {exc}")
    print()
```

```python
import os
from dotenv import load_dotenv
from anthropic import AnthropicFoundry

load_dotenv()

models = [
    {
        "name": "claude-opus-4-5",
        "base_url": "https://project3-docai-resource.services.ai.azure.com/anthropic/",
        "deployment": "claude-opus-4-5",
        "api_key_env": "4.5_OPUS_AZURE_ANTHROPIC_API_KEY",
    },
    {
        "name": "project-04-claude-sonnet-4-6",
        "base_url": "https://9h00200-act-aifoundry.openai.azure.com/anthropic/",
        "deployment": "project-04-claude-sonnet-4-6",
        "api_key_env": "4.6_SONNET_AZURE_ANTHROPIC_API_KEY",
    },
]

prompt = "請用繁體中文，條列整理這段課程心得的重點：競爭觀點、目標設定、執行策略。"
tunable_params = {
    "max_tokens": 512,
    "temperature": 0.3,
    "stop_sequences": [],
}

print("可調參數：")
for key, value in tunable_params.items():
    print(f"- {key}: {value}")
print("注意：temperature 和 top_p 不能同時指定，此模型只支援 temperature")
print()

for item in models:
    api_key = os.getenv(item["api_key_env"])
    if not api_key:
        print(f"=== {item['name']} ===")
        print(f"Error: 找不到 {item['api_key_env']}")
        print()
        continue

    client = AnthropicFoundry(
        api_key=api_key,
        base_url=item["base_url"],
    )

    print(f"=== {item['name']} ===")
    print(f"API key env: {item['api_key_env']}")
    try:
        message = client.messages.create(
            model=item["deployment"],
            messages=[
                {"role": "user", "content": prompt},
            ],
            max_tokens=tunable_params["max_tokens"],
            temperature=tunable_params["temperature"],
            stop_sequences=tunable_params["stop_sequences"],
        )
        print(message.content[0].text)
    except Exception as exc:
        print(f"Error: {exc}")
    print()
```

```python
from pathlib import Path
import os

from dotenv import load_dotenv
from google import genai
from google.genai import types

load_dotenv()

api_key = os.getenv("GEMINI_API_KEY")
if not api_key:
    raise ValueError("找不到 GEMINI_API_KEY，請先確認 .env 已設定")

client = genai.Client(api_key=api_key)

# 官方名稱：
# 1) gemini-3.1-flash-image-preview = Nano Banana 2 Preview
# 2) gemini-2.5-flash-image = Nano Banana
image_model = "gemini-3.1-flash-image-preview"

prompt = (
    "Create a clean premium poster for a Traditional Chinese insurance and health AI platform. "
    "Use a warm ivory background, emerald accents, clear Traditional Chinese headline text, "
    "and a modern professional visual style."
)

output_dir = Path("generated_images")
output_dir.mkdir(exist_ok=True)

response = client.models.generate_content(
    model=image_model,
    contents=[prompt],
    config=types.GenerateContentConfig(
        response_modalities=["TEXT", "IMAGE"],
        image_config=types.ImageConfig(
            aspect_ratio="1:1",
        ),
    ),
)

saved_files = []
parts = response.parts or []

for index, part in enumerate(parts, start=1):
    if getattr(part, "thought", False):
        continue

    if part.text:
        print(f"[TEXT {index}]\n{part.text}\n")
        continue

    if part.inline_data is not None:
        image = part.as_image()
        file_path = output_dir / f"{image_model.replace('.', '_')}_{index}.png"
        image.save(file_path)
        saved_files.append(str(file_path))
        print(f"已儲存圖片：{file_path}")

if not saved_files:
    print("這次回應沒有產生圖片檔。")
else:
    print("測試完成，輸出檔案如下：")
    for file_path in saved_files:
        print(file_path)

# 若想改測 Nano Banana，可把 image_model 改成：
# image_model = "gemini-2.5-flash-image"
```
